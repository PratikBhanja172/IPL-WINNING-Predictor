```python
match_df=match_df[['match_id','city','winner','total_runs']] ### which which columns are needed
```

```python
delivery_df=match_df.merge(delivery,on='match_id')##means you're merging two DataFrames — match_df and delivery — using the match_id column as the common key
```

```python
delivery_df=delivery_df[delivery_df['inning'] == 2]  ### DataFrame to get only the deliveries that occurred in the 2nd innings of the matches
```

```python
delivery_df['current_score'] = delivery_df.groupby('match_id')['total_runs_y'].cumsum()
#You are calculating the running total of runs scored throughout each match, ball by ball.
```

```python
delivery_df
```

```python
delivery_df['runs_left'] = (delivery_df['total_runs_x'] - delivery_df['current_score']).clip(lower=0)
  ### how many runs to be left
```

```python
delivery_df
```

```python
# Total balls in a T20 inning
TOTAL_BALLS = 120

# Calculate balls bowled and left
delivery_df['balls_bowled'] = delivery_df['over'] * 6 + (delivery_df['ball'] - 1)
delivery_df['balls_left'] = (TOTAL_BALLS - delivery_df['balls_bowled']).clip(lower=0)
```

```python
# 1. Create wickets column first
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x: "1" if x != "0" else "0")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype(int)

delivery_df['wickets_fallen'] = delivery_df.groupby('match_id')['player_dismissed'].cumsum()
delivery_df['wickets'] = 10 - delivery_df['wickets_fallen']
```

```python
## crr = runs / overs
delivery_df['crr']=(delivery_df['current_score']*6)/(120-delivery_df['balls_left'])
```

```python
##rrr
delivery_df['rrr']=(delivery_df['runs_left']*6)/delivery_df['balls_left']
```

```python
# This function checks if the batting team in the current row is the same as the match winner.
# If the batting team won the match, it returns 1; otherwise, it returns 0.
# This is useful for analyzing or predicting whether the current batting team won the match.
def result(row): 
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery_df.apply(result,axis=1)
```

```python
# Applies the 'result' function to each row of the DataFrame.
# Adds a new column 'result' where 1 means the batting team won the match, and 0 means they lost.
delivery_df['result'] = delivery_df.apply(result, axis=1)
```

```python
 ### DATA SPLIT
X = final_df.iloc[:, :-1]  # All columns except the last one
y = final_df.iloc[:, -1]   # Only the last column
from  sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer(
    transformers=[
        ('trf', OneHotEncoder(sparse_output=False, drop='first'), ['batting_team', 'bowling_team', 'city'])
    ],
    remainder='passthrough'
)
# ColumnTransformer setup:
# - Applies OneHotEncoding to selected categorical columns: 'batting_team', 'bowling_team', and 'city'.
# - Uses drop='first' to avoid the dummy variable trap by removing the first category.
# - Sets sparse_output=False to return dense arrays.
# - Passes through all other columns (like runs_left, balls_left, etc.) without transformation.
# 
# Why?
# Machine learning models require numerical input.
# This transformer converts categorical data into numerical form (via one-hot encoding),
# while keeping the remaining numerical features as they are.
```

```python
## now vreating pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
pipe=Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])    
```

```python
pipe.fit(X_train,y_train)
# -------------------- Understanding X and y --------------------
# In any machine learning problem:
# X -> Input features (also called independent variables)
#      These are the columns that help the model make predictions.
#      For example: 'batting_team', 'bowling_team', 'city', 
#                   'runs_left', 'balls_left', 'wickets_left', etc.
#
# y -> Target variable (also called dependent variable)
#      This is the column we are trying to predict.
#      For this IPL win probability project, 'y' might represent
#      whether the batting team won or not (0 or 1), or a probability.
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
def match_progression(x_df,match_id,pipe):
    match = x_df[x_df['match_id'] == match_id]
    match = match[(match['ball'] == 6)]
    temp_df = match[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','crr','rrr']].dropna()
    temp_df = temp_df[temp_df['balls_left'] != 0]
    result = pipe.predict_proba(temp_df)
    temp_df['lose'] = np.round(result.T[0]*100,1)
    temp_df['win'] = np.round(result.T[1]*100,1)
    temp_df['end_of_over'] = range(1,temp_df.shape[0]+1)
    
    target = temp_df['total_runs_x'].values[0]
    runs = list(temp_df['runs_left'].values)
    new_runs = runs[:]
    runs.insert(0,target)
    temp_df['runs_after_over'] = np.array(runs)[:-1] - np.array(new_runs)
    wickets = list(temp_df['wickets'].values)
    new_wickets = wickets[:]
    new_wickets.insert(0,10)
    wickets.append(0)
    w = np.array(wickets)
    nw = np.array(new_wickets)
    temp_df['wickets_in_over'] = (nw - w)[0:temp_df.shape[0]]
    
    print("Target-",target)
    temp_df = temp_df[['end_of_over','runs_after_over','wickets_in_over','lose','win']]
    return temp_df,target
    ##win probability each over by over 
```

```python
temp_df,target = match_progression(delivery_df,74,pipe)
temp_df
## how mauch progression 
## here 74 indicates that match number
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(18,8))
plt.plot(temp_df['end_of_over'],temp_df['wickets_in_over'],color='yellow',linewidth=3)
plt.plot(temp_df['end_of_over'],temp_df['win'],color='#00a65a',linewidth=4)
plt.plot(temp_df['end_of_over'],temp_df['lose'],color='red',linewidth=4)
plt.bar(temp_df['end_of_over'],temp_df['runs_after_over'])
plt.title('Target-' + str(target))
```
